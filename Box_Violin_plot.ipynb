```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load your data into a DataFrame
data = pd.read_csv('docking_plot_2.csv')  # Assuming you have a CSV file with your data

# Create the box plot
plt.figure(figsize=(10, 6))
sns.boxplot(x='Molecule_type', y='Docking_score', data=data)
plt.title('Docking Scores of Inhibitors vs. Activators')
plt.xlabel('Molecule_type')
plt.ylabel('Docking_score')
plt.savefig('docking_scores_boxplot.jpeg', dpi=300)
plt.show()
```

```python
# Create the violin plot
plt.figure(figsize=(10, 6))
sns.violinplot(x='Molecule_type', y='Docking_score', data=data, inner='box')
plt.title('Distribution of Docking Scores: Inhibitors vs. Activators')
plt.xlabel('Molecule Type')
plt.ylabel('Docking Score')
plt.savefig('docking_scores_violinplot.jpeg', dpi=300)
plt.show()
```
